```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Gaurang111/Forest_fire_predicting_app/main/dataset.csv")
```

```python
df
```

```python
df = df.drop("Area", axis=1)
```

```python
import tensorflow as tf
import numpy as np
from tensorflow import lite
```

```python
df["Fire Occurrence"].value_counts()
```

```python
X =df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
X
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(3,)))
model.add(tf.keras.layers.Dense(32, activation="relu"))
model.add(tf.keras.layers.Dense(64, activation="relu"))

model.add(tf.keras.layers.Dense(1 ,activation="sigmoid"))

model.compile(loss = "binary_crossentropy",
              optimizer = tf.keras.optimizers.Adam(),
              metrics=["accuracy"])

model.fit(X,y,epochs=200)
```

```python
result =model.evaluate(X,y)
```

```python
input=np.array([X[0]])
answer=round(model.predict(input)[0][0],2)
answer
```

```python
converter = lite.TFLiteConverter.from_keras_model(model)
tfmodel = converter.convert()
open("/content/drive/MyDrive/Colab Notebooks/forestfire.tflite",'wb').write(tfmodel)
```
